# Data Visualizations of MySQL Database

```python
# Creating a Bar Chart of the DataFrame.
plt.figure(figsize = (18, 7))

chart = %sql SELECT job_id, COUNT(job_id) FROM employees GROUP BY job_id 
chart.bar();
```

```python
# Creating a Pie Chart of the DataFrame.
plt.figure(figsize = (18, 7))

chart = %sql SELECT job_id, COUNT(job_id) FROM employees GROUP BY job_id 
chart.pie();
```
